```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Function to fetch historical data
def fetch_data(ticker, start_date, end_date):
    """
    Fetch historical data for a given index.
    
    :param ticker: Index ticker symbol (e.g., '^NSEI' for Nifty 50, '^BSESN' for Sensex)
    :param start_date: The start date for fetching data
    :param end_date: The end date for fetching data
    :return: DataFrame containing historical data
    """
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

# Function to calculate CAGR
def calculate_cagr(start_value, end_value, years):
    """
    Calculate the Compound Annual Growth Rate (CAGR).
    
    :param start_value: Starting price
    :param end_value: Ending price
    :param years: Duration in years
    :return: CAGR value
    """
    return (end_value / start_value) ** (1 / years) - 1

# Function to plot indices
def plot_indices(nifty_data, sensex_data, nifty_label='Nifty 50 Index', sensex_label='Sensex Index'):
    """
    Plot the adjusted close prices of two indices.
    
    :param nifty_data: DataFrame containing Nifty 50 data
    :param sensex_data: DataFrame containing Sensex data
    :param nifty_label: Label for the Nifty 50 in the plot
    :param sensex_label: Label for the Sensex in the plot
    """
    plt.figure(figsize=(10, 6))
    plt.plot(nifty_data.index, nifty_data['Adj Close'], label=nifty_label, color='blue')
    plt.plot(sensex_data.index, sensex_data['Adj Close'], label=sensex_label, color='red')
    
    # Adding title and labels
    plt.title(f'Nifty 50 vs Sensex Index Growth')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.grid(True)
    plt.legend()
    
    # Show the plot
    plt.show()
```

```python
# Main function to calculate and compare Nifty 50 and Sensex
# Define start and end dates
start_date = '1996-04-22'
today = datetime.today().strftime('%Y-%m-%d')

# Fetch Nifty 50 and Sensex data
nifty_data = fetch_data('^NSEI', start_date, today)
sensex_data = fetch_data('^BSESN', start_date, today)

# Ensure both datasets have actual start dates for the data available
actual_start_date_nifty = nifty_data.index[0]
actual_start_date_sensex = sensex_data.index[0]

print(f"Nifty data starts from: {actual_start_date_nifty.date()}")
print(f"Sensex data starts from: {actual_start_date_sensex.date()}")

# Get the starting and ending values
start_value_nifty = nifty_data['Adj Close'].iloc[0]
end_value_nifty = nifty_data['Adj Close'].iloc[-1]

start_value_sensex = sensex_data['Adj Close'].iloc[0]
end_value_sensex = sensex_data['Adj Close'].iloc[-1]

# Calculate the number of years
end_date = nifty_data.index[-1]
years = (end_date - actual_start_date_nifty).days / 365.25

# Calculate CAGR for Nifty 50 and Sensex
cagr_nifty = calculate_cagr(start_value_nifty, end_value_nifty, years)
cagr_sensex = calculate_cagr(start_value_sensex, end_value_sensex, years)

# Display the CAGR results
print(f"CAGR of Nifty 50 from {actual_start_date_nifty.date()} to {end_date.date()} is {cagr_nifty*100:.2f}%")
print(f"CAGR of Sensex from {actual_start_date_sensex.date()} to {end_date.date()} is {cagr_sensex*100:.2f}%")

# Plot the indices
plot_indices(nifty_data, sensex_data)
```
