# 1. Tokenization with NLTK or spaCy

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

text = "Natural Language Processing with Jupyter is fun!"
tokens = word_tokenize(text)
print(tokens)
```

# 2. Using spaCy for Named Entity Recognition (NER)

```python
import spacy

# Load the pre-trained NLP model
nlp = spacy.load('en_core_web_sm')
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

# Extract entities
for entity in doc.ents:
    print(f"{entity.text}: {entity.label_}")
```

# 3. Sentiment Analysis with TextBlob


```python
from textblob import TextBlob

text = "I love using Jupyter for NLP tasks!"
blob = TextBlob(text)
print(blob.sentiment)
```

# 4. Word Embeddings using Gensim


```python
import gensim.downloader as api
word_vectors = api.load("glove-wiki-gigaword-50")

# Find similar words
print(word_vectors.most_similar("apple"))
```

# 5. BERT for Sentence Embeddings (Hugging Face transformers)


```python
from transformers import pipeline

# Load a pre-trained sentiment analysis pipeline
nlp_pipeline = pipeline("sentiment-analysis")

result = nlp_pipeline("Jupyter notebooks are incredibly useful for NLP!")
print(result)
```

# 6. Word Cloud Visualization


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Natural Language Processing with Jupyter is fun and powerful."
wordcloud = WordCloud().generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
